```python
import pandas as pd

TIMELINE_CSV_PATH='outputs/diffusion_timelines_v2.csv'
df = pd.read_csv(TIMELINE_CSV_PATH)
```

```python
graph = {}

for source, reply in zip(df.sourceTweetID, df.replyTweetID):
    if source not in graph.keys():
        graph[source] = set()
    graph[source].add(reply)
```

```python
graph
```

```python
import json

with open('graph.json', 'w') as f:
    json.dump(graph, f, indent=2, default=str)
```
